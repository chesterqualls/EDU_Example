## Begin work with Assess 
> * We will get a list of valid researchid's from enroll data.
* Then we will import, combine and trim down the assess data.
* And then we will clean up some of the data content for analysis.

```python
#first get a list of the researchid's we have in the enroll table, can't do much if we don't have them
df_researchid_all =pd.read_pickle('pickles/am_researchid_all')

print ("Finished")
df_researchid_all
```

```python
df_assess_cohort = pd.read_csv('CRI Student Data/A Mind_Cohort 1_Assess.csv')

df_assess_comp = pd.read_csv('CRI Student Data/A Mind_Comp 1_Assess.csv')

df_assess_all = pd.concat([df_assess_cohort,df_assess_comp])

del df_assess_cohort
del df_assess_comp

df_assess_all.columns = df_assess_all.columns.str.lower()

#The following schools dropped out of the study
df_assess_all = df_assess_all.loc[df_assess_all['schoolcode'] != 2876]
df_assess_all = df_assess_all.loc[df_assess_all['schoolcode'] != 3648]
df_assess_all = df_assess_all.loc[df_assess_all['schoolcode'] != 2134]
df_assess_all = df_assess_all.loc[df_assess_all['schoolcode'] != 5324]


df_assess_all = pd.merge(df_researchid_all, df_assess_all, how='inner', on='researchid')

#del df_researchid_all

#Only need the following columns from the Assess data
df_assess_all = df_assess_all[['researchid', 'schoolcode','schoolname','schoolyear', 'reportinggrade','standard_met_ind','testtype',
                              'performancelevel','testgrade','scalescore','subject']]

#Change some of the column names for clarity
df_assess_all = df_assess_all.rename(columns={'standard_met_ind': 'met_standard'})


print ("Finished")
df_assess_all
```

```python
#make other modifications to the data.

#drop Nulls
df_assess_all = df_assess_all.dropna(subset=['performancelevel','scalescore'])

#Change standard met to number
df_assess_all['met_standard'] = df_assess_all['met_standard'].map({'Yes': 1, 'No': 0})

print ("Finished")
df_assess_all
```

```python
df_assess_all.info()
```

```python
#Need to convert some columns to numeric
df_assess_all['testgrade'] = pd.to_numeric(df_assess_all['testgrade'], errors='coerce', downcast='integer')
```

```python
#Next need to identify math assessessments is_math


df_assess_all['is_math'] = np.where((df_assess_all['subject'].str.contains('Math')),1,0)
  

print ("Finished")
#print (df_grades_ia)
```
